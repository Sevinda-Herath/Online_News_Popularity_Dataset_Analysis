```python
# Importing the Dataset
df = pd.read_csv("OnlineNewsPopularity/OnlineNewsPopularity.csv")
```

```python
# Removing any leading or trailing whitespace from the column names
df.columns = df.columns.str.strip()
```

```python
# Function to determine the news category
def get_news_category(row):
    if row['data_channel_is_lifestyle']:
        return 'Lifestyle'
    elif row['data_channel_is_entertainment']:
        return 'Entertainment'
    elif row['data_channel_is_bus']:
        return 'Business'
    elif row['data_channel_is_socmed']:
        return 'Social Media'
    elif row['data_channel_is_tech']:
        return 'Tech'
    elif row['data_channel_is_world']:
        return 'World'
    return 'Other'

# Createing the 'news_category' column
df['news_category'] = df.apply(get_news_category, axis=1)
```

```python
# Counting Articles by Category
articles_count = df['news_category'].value_counts()

# Display Article Count
print("Article Count by Category:")
print(articles_count)

# Plotting article count by category
plt.figure(figsize=(8, 5))
articles_count.plot(kind='bar', color='lightblue')
plt.title('Number of Articles by News Category')
plt.xlabel('News Category')
plt.ylabel('Number of Articles')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Group by news category and sum the shares
shares_by_category = df.groupby('news_category')['shares'].sum()

# Display the shares by category
print("Shares by News Category:")
print(shares_by_category)

# Plotting the shares by news category
plt.figure(figsize=(8, 5))
shares_by_category.plot(kind='bar', color='lightcoral')
plt.title('Total Shares by News Category')
plt.xlabel('News Category')
plt.ylabel('Total Shares')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Function to determine publication day
def get_publication_day(row):
    if row['weekday_is_monday']:
        return 'Monday'
    elif row['weekday_is_tuesday']:
        return 'Tuesday'
    elif row['weekday_is_wednesday']:
        return 'Wednesday'
    elif row['weekday_is_thursday']:
        return 'Thursday'
    elif row['weekday_is_friday']:
        return 'Friday'
    elif row['weekday_is_saturday']:
        return 'Saturday'
    elif row['weekday_is_sunday']:
        return 'Sunday'
    return None

# Apply the function to create new columns
df['publication_day'] = df.apply(get_publication_day, axis=1)
```

```python
# Group by publication day and sum shares
shares_by_day = df.groupby('publication_day')['shares'].sum().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# Display Shares by Publication Day
print("Shares by Publication Day")
print(shares_by_day)

# Bar chart for shares by publication day
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
shares_by_day.plot(kind='bar', color='skyblue')
plt.title('Shares by Publication Day')
plt.xlabel('Publication Day')
plt.ylabel('Total Shares')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Function to Determine Sentiment Category
def get_sentiment_category(polarity):
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Apply the function to create new columns
df['SentimentCategory'] = df['global_sentiment_polarity'].apply(get_sentiment_category)
```

```python
# Group by SentimentCategory and sum the shares
shares_by_sentiment = df.groupby('SentimentCategory')['shares'].sum()

print("Shares by Sentiment Category:")
print(shares_by_sentiment)

# Plotting the shares by sentiment category
plt.figure(figsize=(8, 5))
shares_by_sentiment.plot(kind='bar', color=['red', 'blue', 'green']) 
plt.title('Total Shares by Sentiment Category')
plt.xlabel('Sentiment Category')
plt.ylabel('Total Shares')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
